# Transmission lines 

Whenever we need to transfer a signal over a cable or a transmission line, we always talk about impedance matching. So many times I have heard that this is a 50 Ohm coaxial cable and we need to match impedance device output impedance and device input impedance. A 50-ohm coaxial cable is commonly used in signal transmission in a research environment. A lot of scientists are trained to look for output impedance and make sure it is 50 ohm one. Then we look for input impedance and ensure that it is matched to 50 ohms as well. And the last piece is the transfer line, which commonly a 50-ohm coaxial cable one can find in every research laboratory that deals with triggers and measurements.

Recently I had a small project where I wanted to build a simple passive splitter and delay. The idea was to take one signal, split it into two arms, delay one of the arms and send it into a differential input of a device at the end of the transmission line. 

Before we jump into different ways we can achieve the task. Let us discuss how we are going to measure and visualize what is going on at different parts of the transmission line. First, I am going to use GW Instek SFG-2010 pulse generator to create a pulse train of defined width. I will be interested in the rising edge of a square pulse. It is a good way to define a specific time. Figure 1 shows the rising edge of a pulse generated by the pulse generator at 100 kHz and panel b shows the zoomed-in version. It is a very nice and clean pulse going from 0 to 2V.
